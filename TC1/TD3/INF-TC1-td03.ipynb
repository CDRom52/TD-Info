## Exercice 1 : Introduction aux arbres binaires

Dans cet exercice vous allez créer et parcourir un arbre binaire. Un arbre binaire est une structure hiérarchique qui possède les propriétés suivantes :

- il est dit _binaire_ car il comporte des nœuds ayant au _maximum_ deux enfants
- il est dit _complet_ si tous les niveaux de l'arbre sont remplis (sauf éventuellement le dernier), et dans lequel les feuilles sont alignées à gauche
- il est dit _équilibré_ si la différence de hauteur entre les sous-arbres gauche et droit de chaque nœud est au plus 1

Voici un exemple d'arbre binaire :

```
          1
         / \
        2   3
       / \ / \
      4  5 6  7
```

Dans cet exercice nous stockeront l'arbre binaire avec une structure de données _explicite_ (en POO comme ci-dessous, et non plus dans un tableau comme pour le tas dans le TD 2) qui contient des valeurs entières (et uniques, à savoir deux nœuds n'auront pas la même valeur `value`) dans chaque noeud :

````python
class Node:
    def __init__(self, value: int, left = None, right = None):
        self.value = value
        self.left = left
        self.right = right
````

**Question 1.2** Proposez une méthode `bfs` de _parcours en largeur_ de l'arbre afin d'afficher la valeur des nœuds dans l'ordre croissant. Pour cela vous utiliserez une structure de données de File (sous forme de liste, cela sera suffisant). Une méthode possible pour mener cela à bien consiste à :

1. intialiser la file avec la racine de l'arbre ;
2. dé-filer une valeur et la stocker dans une liste de résultats ;
3. en-filer ses enfants (si il y en a) dans la file ;
4. répéter l'étape 2 jusqu'à ce que la file soit vide, renvoyer le résultat.

````python
def bfs(node):
    # YOUR CODE HERE
    queue = [node]
    res = []
    while queue:
        current = queue.pop(0)
        res.append(current.value)
        if current.left:
            queue.append(current.left)
        if current.right:
            queue.append(current.right)
    return res
````

````python
assert bfs(root) == [1, 2, 3, 4, 5, 6, 7]
````

**Question 1.3** - Écrivez une fonction `depth` de calcul de la _profondeur_ d'un noeud d'un arbre. La profondeur est définie comme la distance entre ce nœud et la racine (celle-ci aura une profondeur de `0`). Autrement dit la profondeur correspond au nombre d'arêtes entre un noeud et la racine de l'arbre. Cette fonction prendra en paramètre la racine de l'arbre `root`, ainsi que le noeud dont on veut calculer la profondeur avec le paramètre `target_value`.

Conseils : 

- s'inspirer de la fonction précédente de parcours en largeur en incluant la profondeur de chaque noeud parcouru lors de son ajout dans la file;
- autrement dit rajouter un tuple `(noeud, profondeur)` au lieu du simple noeud parcouru afin de mémoriser la profondeur des nœuds parcourus.

````python
def depth(root: Node, target_value: int) -> int:
    # YOUR CODE HERE
    queue = [(root, 0)]
    while queue:
        current, current_depth = queue.pop(0)
        if current.value == target_value:
            return current_depth
        if current.left:
            queue.append((current.left, current_depth + 1))
        if current.right:
            queue.append((current.right, current_depth + 1))
    return -1
````

````python
assert depth(root, 1) == 0
assert depth(root, 2) == 1
assert depth(root, 3) == 1
assert depth(root, 4) == 2
````

**Question 1.4** - Écrivez une fonction `height` de calcul de la _hauteur_ d'un arbre définie comme la profondeur maximale possible dans un arbre. Vous pouvez l'implémenter comme la profondeur maximale des nœuds de l'arbre, ou bien de manière récursive.

````python
def height(root: Node) -> int:
    # YOUR CODE HERE
    if root is None:
        return 0
    else:
        left_height = height(root.left)
        right_height = height(root.right)
        return max(left_height, right_height) + 1
````

````python
assert height(root) == 3
````

**Question 1.5** - Écrivez une fonction de validation afin de déterminer si un Arbre est bien équilibré, autrement dit s'il n'y a pas de différence de profondeur supérieur à 1 entre les feuilles d'un arbre. Conseil : utilisez une approche récursive.

````python
def est_equilibre(root: Node) -> bool:
    # YOUR CODE HERE
    if root is None:
        return True, 0
    left_balanced, left_height = est_equilibre(root.left)
    right_balanced, right_height = est_equilibre(root.right)
    balanced = left_balanced and right_balanced and abs(left_height - right_height) <= 1
    height = max(left_height, right_height) + 1
    return balanced, height
````

````python
assert est_equilibre(root)
````

## Exercice 2 : Arbres syntaxiques

Un Arbre _syntaxique_ permet le stockage d'une expression structurée, par exemple une équation. Dans cet exercice nous allons modéliser un tel arbre sous forme d'arbre binaire (exemple ci-dessous à gauche) afin de réaliser un calcul arithmétique simple à partir de l'expression fournie de manière textuelle :

Expression : `( 3 - 2 ) * ( 7 + ( 10 / 2 ) )`

Résultat : `12.0`

Nous ferons l'hypothèse que les opérations sont limitées à `+ - / *`, seront toujours binaires et porteront sur des valeurs numériques entières (mais le résultat peut ne pas être un entier).

```
        *
       / \
      -   +
     / \ / \
    3  2  7  /
            / \
           10  2
```

Vous utiliserez la structure d'arbre binaire ci-dessous afin de stocker l'arbre comme pour l'abre binaire précédent :

````python
class Noeud:
    def __init__(self, v = None, g = None, d = None) -> None:
        self.valeur = v
        self.gauche = g
        self.droit = d
````

**Question 2.1** - Stockez l'arbre syntaxique donné en exemple graphique en utilisant la structure de données d'arbre binaire ci-dessus (il s'agit d'écrire une instance d'arbre dans la variable `arbre`, et non une fonction).

````python
# YOUR CODE HERE
arbre = Noeud('*', Noeud('-', Noeud(3), Noeud(2)), Noeud('+', Noeud(7), Noeud('/', Noeud(10), Noeud(2))))
````

````python
assert arbre.valeur == "*"
assert arbre.gauche.valeur == "-"
assert arbre.droit.valeur == "+"
````

**Question 2.2** - Implémentez désormais une méthode d'évaluation automatique d'un arbre syntaxique tel que vous l'avez stocké dans la variable `arbre` ci-dessus. 

Conseil : proposez une solution récursive avec un cas d'arrêt et des appels récursifs comme suit :
- si la valeur du noeud en cours est un opérateur, l'appliquer sur les deux sous-branches;
- si il s'agit d'une valeur numérique, renvoyez cette valeur (cas d'arrêt car c'est une feuille de l'arbre).

````python
def eval(r: Noeud) -> float:
    # YOUR CODE HERE
    if r.valeur == '+':
        return eval(r.gauche) + eval(r.droit)
    elif r.valeur == '-':
        return eval(r.gauche) - eval(r.droit)
    elif r.valeur == '*':
        return eval(r.gauche) * eval(r.droit)
    elif r.valeur == '/':
        return eval(r.gauche) / eval(r.droit)
    else:
        return r.valeur
````

````python
assert eval(arbre) == 12.0
````

**Question 2.3** - Écrivez une méthode permettant de construire l'arbre à partir d'une expression fournie sous forme de chaîne de caractères en entrée comme `( ( 3 - 2 ) * ( 7 + ( 10 / 2 ) ) )`. Les espaces sont nécessaires et vous permettront de faire un `.split(" ")`. 

Conseil : 
- parcourez caractère par caractère l'expression textuelle transformée en liste;
- créez au fur et à mesure l'arbre initialisé par une racine et une variable stockage le noeud courant en cours de construction;
- identifiez les 4 cas possibles : début d'une expression, valeurs numériques, opérateur, fin d'une expression :
    - si une expression commence, créez un nouveau fils gauche au noeud courant et stockez le noeud courant dans une Pile
    - si une expression finie, considérez comme noeud courant l'élément que vous dé-Pilerez
    - si un opérateur, stockez sa valeur dans le noeud courant, empilez ce nœud mais et rajoutez un fils droit qui devient le noeud courant
    - si une valeur numérique, la stocker dans le noeud courant et considérez comme noeud courant l'élément que vous dé-Pilerez

````python
def construit_arbre(exp: str) -> Noeud:
    # YOUR CODE HERE
    pile = []
    racine = Noeud()
    courant = racine
    for el in exp.split():
        if el == '(':
            nouveau_noeud = Noeud()
            if courant.gauche is None:
                courant.gauche = nouveau_noeud
            else:
                courant.droit = nouveau_noeud
            pile.append(courant)
            courant = nouveau_noeud
        elif el == ')':
            courant = pile.pop()
        elif el in ['+', '-', '*', '/']:
            courant.valeur = el
            pile.append(courant)
            nouveau_noeud = Noeud()
            courant.droit = nouveau_noeud
            courant = nouveau_noeud
        else:  # el is a number
            courant.valeur = int(el)
            courant = pile.pop()
    return racine
````

````python
r = construit_arbre("( ( 3 - 2 ) * ( 7 + ( 10 / 2 ) )")
````

Enfin vérifiez que vous obtenez la bonne valeur en évaluant l'expression.

````python
assert eval(r) == 12.0
````

**Question 2.4 (Bonus) -** Ecrire une fonction qui renvoie `True` ou `False` si l'arbre est bien un arbre syntaxique tel que défini en introduction. Autrement dit qu'il est binaire, et comporte des opérateurs partout sauf aux feuilles.

Proposez d'abord une méthode itérative :

````python
def valide_iter(r: Noeud) -> bool:
    # YOUR CODE HERE
    if r is None:
        return True
    pile = [r]
    while pile:
        current = pile.pop()
        if current.gauche is None and current.droit is None:
            if not isinstance(current.valeur, int):
                return False
        else:
            if current.valeur not in ['+', '-', '*', '/']:
                return False
            if current.gauche:
                pile.append(current.gauche)
            if current.droit:
                pile.append(current.droit)
    return True
````

````python
assert valide_iter(arbre)
````

Proposez ensuite une méthode récursive :

````python
def valide_rec(r: Noeud) -> bool:
    # YOUR CODE HERE
    if r is None:
        return True
    if r.gauche is None and r.droit is None:
        return isinstance(r.valeur, int)
    if r.valeur not in ['+', '-', '*', '/']:
        return False
    return valide_rec(r.gauche) and valide_rec(r.droit)
````

````python
assert valide_rec(arbre)
````

**Question 2.5 (Bonus) -** Créez une table de hachage afin de mémoriser les opérations déjà réalisées pour l'arbre syntaxique. Conseil : utilisez les informations sur le noeud comme clé (valeurs, opérateur).

````python
def eval_hash(r: Noeud, memo = None) -> float:
    # YOUR CODE HERE
    if memo is None:
        memo = {}
    key = (r.valeur, id(r.gauche), id(r.droit))
    if key in memo:
        return memo[key]
    if r.valeur == '+':
        result = eval_hash(r.gauche, memo) + eval_hash(r.droit, memo)
    elif r.valeur == '-':
        result = eval_hash(r.gauche, memo) - eval_hash(r.droit, memo)
    elif r.valeur == '*':
        result = eval_hash(r.gauche, memo) * eval_hash(r.droit, memo)
    elif r.valeur == '/':
        result = eval_hash(r.gauche, memo) / eval_hash(r.droit, memo)
    else:
        result = r.valeur
    memo[key] = result
    return result
````

````python
assert eval_hash(arbre) == eval(arbre)
````
